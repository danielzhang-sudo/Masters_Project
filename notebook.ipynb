```python
import pandas as pd

df = pd.read_csv('/content/lending_club_loan_two.csv')
pd.set_option('display.max_columns', None)
```

```python
df.nunique()
```

```python
# discard
# address

# ordinal encoding
# grade, sub_grade, emp_length, home_ownership, verification_status, earliest_cr_line,

# numerical encoding:
# loan_amount, term, int_rate, installment, annual_inc, issue date, purpose, dti,
# open_acc, pub_rec, revol_bal, revol_util, total_acc, initial_list_status, application_type, mort_acc, pub_rec_bankruptcies

# word embedding
# emp_title, title

# class
# loan_status
```

```python
df_copy = df.copy()
df_copy.drop('address', axis=1, inplace=True)
```

```python
df_copy = pd.get_dummies(df_copy, columns=['home_ownership', 'verification_status', 'purpose', 'initial_list_status', 'application_type', 'loan_status'], dtype=int)
df_copy.head()
```

```python
df_copy['grade'] = pd.Categorical(list(df_copy['grade'])).codes
df_copy['sub_grade'] = pd.Categorical(list(df_copy['sub_grade'])).codes
```

```python
df_copy['emp_length'] = df_copy['emp_length'].fillna(value='nan')
```

```python
# Define the mapping of emp_length values to their ordinal encoding
mapping = {'nan': -1, '< 1 year': 11, '1 year': 10, '2 years': 9, '3 years': 8, '4 years': 7, '5 years': 6, '6 years': 5, '7 years': 4, '8 years': 3, '9 years': 2, '10+ years': 1}

# Apply the mapping to the 'emp_length' column
df_copy['emp_length'] = df_copy['emp_length'].map(mapping)

df_copy.head()
```

```python
emp_title = list(df_copy['emp_title'])
```

```python
import gensim.downloader as api

# Download pre-trained Word2Vec model
model = api.load('word2vec-google-news-300')

# Define a function to encode words into embeddings
def encode_words(words):
    embeddings = []
    for word in words:
        try:
            embedding = model[word]
            embeddings.append(embedding)
        except KeyError:
            # Word not found in vocabulary, use zero vector
            embeddings.append([0]*model.vector_size)
    return embeddings

# Example words to encode
words = ['cat', 'dog', 'apple']

# Encode words into embeddings
word_embeddings = encode_words(words)

print(word_embeddings)
```

```python
len(word_embeddings[1])
```

```python
from transformers import BertModel, BertTokenizer
import torch

# Load pre-trained BERT model and tokenizer
model = BertModel.from_pretrained('bert-base-uncased')
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Encode the sentence
sentence = "good morning friend"
input_ids = tokenizer.encode(sentence, add_special_tokens=True, return_tensors='pt')

# Get the BERT embedding
with torch.no_grad():
    outputs = model(input_ids)
    embedded_sentence = outputs[0][:, 0, :].numpy()

print(embedded_sentence)
```

```python
len(embedded_sentence[0])
```

```python
# Encode the sentence
sentences = list(df_copy['emp_title'])
print(len(sentences))
print(sentences[0])
```

```python
sentences_short = sentences[:10]
```

```python
input_idss = []
for sentence in sentences_short:
  input_idss.append(tokenizer.encode(sentence, add_special_tokens=True, return_tensors='pt'))

embedded_sentences = []

# Get the BERT embedding
with torch.no_grad():
  for input_ids in input_idss:
    outputs = model(input_ids)
    embedded_sentences.append(outputs[0][:, 0, :].numpy())
```

```python
for i in range(10):
  print(len(embedded_sentences[i][0]))
```
